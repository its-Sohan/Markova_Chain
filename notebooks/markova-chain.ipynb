```python
import numpy as np 
```

```python
P = np.array([
    [0.3, 0.3, 0.2, 0.1, 0.05, 0.05], 
    [0.2, 0.4, 0.3, 0.05, 0.0, 0.05],
    [0.2, 0.3, 0.4, 0.0, 0.0, 0.1], 
    [0.3, 0.0, 0.0, 0.4, 0.1, 0.2],
    [0.1, 0.1, 0.0, 0.1, 0.6, 0.1], 
    [0.1, 0.3, 0.3, 0.1, 0.0, 0.2]
])
```

```python
# starting at state 1
pi_0 = np.array([1,0,0,0,0,0])

steps = 3

#store probability vectors
history = [pi_0]

# Multiply iteratively: π_n = π_0 × P^n
pi = pi_0.copy()
for _ in range(steps):
    pi = pi @ P   # matrix multiplication
    history.append(pi)
history= np.array(history)
```
