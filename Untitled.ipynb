```python
import numpy as np

# Define the states
states = ['Sunny', 'Cloudy', 'Rainy']
state_map = {state: i for i, state in enumerate(states)}
print(state_map)
```

```python
# Matrix for step 1 -> 2: Tends to stay sunny
P1 = np.array([
    [0.8, 0.15, 0.05], # From Sunny
    [0.3, 0.6, 0.1],   # From Cloudy
    [0.2, 0.4, 0.4]    # From Rainy
])

# Matrix for step 2 -> 3: Higher chance of getting cloudy
P2 = np.array([
    [0.5, 0.4, 0.1],   # From Sunny
    [0.1, 0.7, 0.2],   # From Cloudy
    [0.1, 0.5, 0.4]    # From Rainy
])

# Matrix for step 3 -> 4: Higher chance of rain
P3 = np.array([
    [0.4, 0.4, 0.2],   # From Sunny
    [0.2, 0.4, 0.4],   # From Cloudy
    [0.1, 0.3, 0.6]    # From Rainy
])
```

```python
initial_state = np.array([0 , 0 ,  1])
state_history = [initial_state]

for P in [P1, P2, P3]: 
    initial_state = initial_state @ P 
    state_history.append(initial_state)

    # Display results
weather_states = ['Sunny', 'Cloudy', 'Rainy']
for t, s in enumerate(state_history, start=0):
    print(f"Time {t}:")
    for weather, prob in zip(weather_states, s):
        print(f"{weather}: {prob:.2f}")
```
