```python
import numpy as np 
```

```python
#transition matrix 
P = np.array([
    [0.4, 0.4, 0.2], 
    [0.1, 0.7, 0.2],
    [0.5, 0.4, 0.1]
])
```

```python
# starting at state 1
pi_0 = np.array([1, 0, 0])

steps = 20

#store probability vectors
history = [pi_0]

# Multiply iteratively: π_n = π_0 × P^n
pi = pi_0.copy()
for _ in range(steps):
    pi = pi @ P   # matrix multiplication
    history.append(pi)
```

```python
print(history)
```
